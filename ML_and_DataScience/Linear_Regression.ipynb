```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([7,14,15,18,19])
n = np.size(x)

x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean, y_mean)
```

```python
ssy = np.sum(x*y) - n*x_mean*y_mean
ssx = np.sum(x*x) - n*x_mean*x_mean
m_slope = ssy/ssx
c_intercept = y_mean - m_slope*x_mean 
print(m_slope, c_intercept)
```

```python
y_pred = m_slope*x + c_intercept
```

```python
plt.plot(x,y_pred)
plt.xlabel("x Independent variable")
plt.ylabel("y dependent variable")
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.xlabel("x Independent variable")
plt.ylabel("y dependent variable")
```

```python
err = y- y_pred
sq_err = np.sum(err**2)
mean_sq_err = sq_err/n
root_mean_sq_err = np.sqrt(mean_sq_err)
sst = np.sum((y-y_mean)**2)
r2_sc = 1- (sq_err/sst)
print(sq_err, mean_sq_err, root_mean_sq_err, r2_sc)
```

```python
x = x.reshape(-1,1)
regression_model = LinearRegression()
regression_model.fit(x,y)
```

```python
y_pred = regression_model.predict(x)
mean_sq_err = mean_squared_error(y, y_pred)
root_mean_sq_err = np.sqrt(mean_squared_error(y, y_pred))
r2_sc = r2_score(y, y_pred)
print("Coefficient: ", regression_model.coef_)
print("intercept : ", regression_model.intercept_)
print("Mean squared err, Root mean squared err, r2 score : ", mean_sq_err, root_mean_sq_err, r2_sc)
```
