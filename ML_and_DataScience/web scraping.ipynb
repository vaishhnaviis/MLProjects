```python
url = 'https://www.geeksforgeeks.org/python-programming-language/'
r= requests.get(url)
soup = bs(r.content, 'html.parser')
```

```python
images= soup.select('img')
print(images)
images_list =[]
```

```python
for image in images:
    src = image.get('src')
    alt = image.get('alt')
    images_list.append({'src': src, 'alt': alt})
```

```python
for image in images_list:
    print(image)
```
